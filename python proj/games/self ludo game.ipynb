```python
import random

class LudoGame:
    def __init__(self):
        self.board = [0] * 52
        self.players = ["Player 1", "Player 2", "Player 3", "Player 4"]
        self.current_player = 0

    def roll_dice(self):
        return random.randint(1, 6)

    def move_piece(self, player, steps):
        current_position = self.board[player]
        new_position = (current_position + steps) % 52
        self.board[player] = new_position

    def play_turn(self):
        player = self.current_player
        input(f"{self.players[player]}, press Enter to roll the dice...")
        
        dice_result = self.roll_dice()
        print(f"{self.players[player]} rolled a {dice_result}.")

        self.move_piece(player, dice_result)
        self.display_board()

        if self.check_winner():
            print(f"Congratulations, {self.players[player]}! You won!")
            exit()

        self.current_player = (self.current_player + 1) % len(self.players)

    def check_winner(self):
        return all(position == 51 for position in self.board)

    def display_board(self):
        print("\nCurrent Board:")
        for i, player in enumerate(self.players):
            print(f"{player}: {self.board[i]}")
        print("\n")


def main():
    ludo_game = LudoGame()

    while True:
        ludo_game.play_turn()


if __name__ == "__main__":
    main()
```
